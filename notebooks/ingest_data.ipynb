```python
from utils.postgres_utils import PostrgresService
import datetime
import json

pg_service = PostrgresService('utils/secrets.json')
```

```python
query = """
select distinct on (id) * from metabase.features_preselected_v2
"""
```

```python
data_iterator = pg_service.query_data(query, batch_size=1000)
```

```python
with open('data_dump_full.csv', 'a') as f:
    c=0
    for item in data_iterator:
        final = {}
        for k, v in item.items():
            if isinstance(v, datetime.datetime):
                final[k] = str(v)
            else:
                final[k] = v
        f.write(json.dumps(final)+'\n')
        c+=1
        if c%1000 == 0:
            print(c)
```
